```python
# Importing libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# Reading image

image = cv2.imread('Bird_Image.jpg')
```

### Original image

```python
plt.figure(figsize=(10,15))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB), cmap = 'gray')
plt.show()
```

### Image convert to grayscale version

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,15))
plt.imshow(gray,cmap = 'gray')
plt.show()
```

### Image convert to inverted version

```python
inverted_image  = 255 - gray
plt.figure(figsize = (10, 15))
plt.imshow(inverted_image,cmap='gray')
plt.show()
```

### Image convert to blur inverted version

```python
blurred_image = cv2.GaussianBlur(inverted_image, ksize=(21,21), sigmaX=0,sigmaY=0)
plt.figure(figsize=(10,15))
plt.imshow(blurred_image,cmap = 'gray')
plt.show()
```

### Image converted to pencil sketch version

```python
pencil_sketch = cv2.divide(gray,255-blurred_image,scale = 256)
plt.figure(figsize=(10,15))
plt.imshow(pencil_sketch,cmap = 'gray')
plt.show()
```
